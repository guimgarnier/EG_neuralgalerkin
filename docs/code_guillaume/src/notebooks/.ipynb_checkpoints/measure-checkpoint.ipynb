# Metropolis Hastings Sampler Algorithm

In this Notebook, we describe and we implement with Python the Metropolis Hastins Sampler Algorithm. 

## Description of the MH Algorithm

The Metropolis-Hastings algorithm is a Markov chain Monte Carlo (MCMC) method used for sampling from a probability distribution $f$ that is difficult to sample directly. 

<div class="alert alert-block alert-info"> <b>Key Point - </b> The principle is to approximate the target distribution by a Markov chain that converges to the target distribution. </div>

*   <b>Step 1</b> : Choose an initiage point $x_0$ in the support of $f$;
*   <b>Step 2</b> : Draw a candidate $x'$ from the law $Q(y | x)$;
*   <b>Step 3</b> : Compute $\alpha = \frac{f(x')}{f(x_0)}$;
*   <b>Step 4</b> : Draw $U$ from the uniform density $\mathcal{U}(0,1)$; 
*   <b>Step 5</b> : 
    * If $U < \alpha$, then reject $x_1$ and set $x_1 = x_0$;
    * Else, set $x_1 = x'$;
*   <b>Step 6</b> : Come back to step 2 until you have any points $(x_n)_n$.



## Code of the MH Algorithm

```python
import torch
import torch.distributions as dist
torch.manual_seed(0)


class MetropolisHastingsSampler:
    def __init__(self, num_samples, initial_state, target_distribution, proposal_distribution):
        self.num_samples = num_samples
        self.initial_state = initial_state
        self.target_distribution = target_distribution
        self.proposal_distribution = proposal_distribution

    def metropolis_hastings(self):
        samples = torch.empty((self.num_samples, self.initial_state.size(0)))
        current_state = self.initial_state
        accepted_samples = 0

        for i in range(self.num_samples):
            # Sample a candidate from the proposal distribution
            candidate_state = self.proposal_distribution(current_state).sample()

            # Calculate the acceptance ratio
            accept_ratio = (self.target_distribution(candidate_state)) / (self.target_distribution(current_state))

            # Accept or reject the candidate
            if torch.rand(1) < accept_ratio:
                current_state = candidate_state
                accepted_samples += 1

            samples[i] = current_state

        acceptance_rate = accepted_samples / self.num_samples
        return samples, acceptance_rate
```

# Example in 1D

We generate 5000 samples from $\mathcal{N}(0,1)$ bu the MH algorithm. Here, we choose the initial point $x_0 = 0$ and the proposal distribution $\mathcal{N}(0, 0.5)$. 

```python
# Number of samples to generate
num_samples = 5000
dimension_samples = 1

    # Initial state of the Markov chain
initial_state = torch.zeros(dimension_samples)

    # Target distribution : normal_law
def target_distribution(x):
    """Target distribution: a standard Gaussian."""
    return torch.exp(-0.5 * x**2) / torch.sqrt(2 * torch.tensor(3.141592653589793))

    # Proposal distribution: a normal law with mean x and standard deviation 'sigma'
def proposal_distribution(x):
    """Proposal distribution: a Gaussian with mean x and standard deviation sigma."""
    sigma = 0.5
    return dist.Normal(x, sigma)

    # Create the sampler instance
sampler = MetropolisHastingsSampler(num_samples, initial_state, target_distribution, proposal_distribution)
    # Run Metropolis-Hastings algorithm
samples, acceptance_rate = sampler.metropolis_hastings()

###########################################
# Plot the histogram and the real density #

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

plt.hist(samples.detach().numpy(), bins=20, density=True)

x_axis = np.arange(-10, 10, 0.001)
plt.plot(x_axis, norm.pdf(x_axis, 0, 1,))

plt.show()
```

```python
# Number of samples to generate
num_samples = 5000
dimension_samples = 1

    # Initial state of the Markov chain
initial_state = torch.zeros(dimension_samples)

def target_distribution(x):
    """Target distribution: a exponential law with intensity lamb."""
    lamb = 0.6
    if x < 0:
      return 0
    else:
      return lamb * torch.exp(-lamb * x)

    # Proposal distribution: a normal law with mean x and standard deviation 'sigma'
def proposal_distribution(x):
    """Proposal distribution: a Gaussian with mean x and standard deviation sigma."""
    sigma = 0.5
    return dist.Normal(x, sigma)

    # Create the sampler instance
sampler = MetropolisHastingsSampler(num_samples, initial_state, target_distribution, proposal_distribution)
    # Run Metropolis-Hastings algorithm
samples, acceptance_rate = sampler.metropolis_hastings()



###########################################
# Plot the histogram and the real density #

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon

plt.hist(samples.detach().numpy(), bins=20, density=True)

x_axis = np.arange(-10, 10, 0.001)
plt.plot(x_axis, expon.pdf(x_axis, scale=1/0.6,))

plt.show()
```

# Example in 2D

```python
    # Number of samples to generate
num_samples = 5000
dimension_samples = 2

    # Initial state of the Markov chain
initial_state = torch.zeros(dimension_samples)

    # Target distribution: a bivariate Gaussian
def target_distribution(x):
  mean = torch.tensor([2.0, -1.0])
  covariance_matrix = torch.tensor([[1.0, 0.7], [0.7, 1.5]])
  mv_normal = dist.MultivariateNormal(mean, covariance_matrix)
  return torch.exp(mv_normal.log_prob(x))

    # Proposal distribution: a bivariate Gaussian with mean x and standard deviation 'sigma'
def proposal_distribution(x):
  sigma = 0.5
  return dist.MultivariateNormal(x, torch.eye(x.size(0)) * sigma)

    # Create the sampler instance
sampler = MetropolisHastingsSampler(num_samples, initial_state, target_distribution, proposal_distribution)

    # Run Metropolis-Hastings algorithm
samples, acceptance_rate = sampler.metropolis_hastings()
```

Some informations

```python
# Display results
print(f"Acceptance rate: {acceptance_rate:.2f}")
print(f"Empirical mean: {samples.mean(dim=0)}")
print(f"Empirical covariance: \n{torch.cov(samples.T)}")
```

# Compute a Monte Carlo Integral

```python
import torch
import torch.distributions as dist

def monte_carlo_integration(func, sample):
    """
    Compute the integral of a given function using the Monte Carlo method.

    Parameters:
        func (callable): The function to integrate. It should accept a tensor input of the form 
          tensor([[x1, y1, z1],
                  [x2, y2, z2],
                  ...,
                  [x3, y3, z3],])
          and give an output tensor of the form tensor([value_sample1, value_sample2, ..., value_sampleN])
        samples(tensor): Random sample where integrate

    Returns:
        torch.Tensor: Approximated integral value.
    """
    # Evaluate the function at the generated samples
    func_values = func(samples)

    # Compute the approximate integral
    integral = torch.mean(func_values)
    return integral
```

Example en 1D
$$\int_0^1 (sin(x) + x^2) dx$$

```python
# Define the function to integrate (e.g., sin(x) + x**2)
def target_function(x):
  return (torch.sin(x) + x**2)

# Number of random samples for Monte Carlo integration
num_samples = 10000
dim_samples = 1

# Generate random samples within the integration domain
uniform_dist = dist.uniform.Uniform(0, 1)
samples = uniform_dist.sample(torch.Size([dim_samples, num_samples]))

    # Compute the integral
result = monte_carlo_integration(target_function, samples)
print("Approximated integral:", result.item())
```

Example in 2D
$$ \int_{[0,1]^2} (x^2 + y^2) dxdy$$

```python
# Define the function to integrate (e.g. x^2 + y^2)
def target_function(x):
  target = torch.zeros(x.size(0))
  for iter in range(x.size(0)):
    target[iter] = x[iter,0]**2 + x[iter,1]**2 
  return target

num_samples = 10000
dim_samples = 2

# Generate random samples within the integration domain
uniform_dist = dist.uniform.Uniform(0, 1)
samples = uniform_dist.sample(torch.Size([dim_samples,num_samples])).T

result = monte_carlo_integration(target_function, samples)
print("value: ", result)
```
