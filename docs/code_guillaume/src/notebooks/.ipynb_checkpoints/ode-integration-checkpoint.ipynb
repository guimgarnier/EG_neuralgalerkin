```python
import torch
from torchdiffeq import odeint
import matplotlib.pyplot as plt
```

# The Lorentz attractor with torchdiffeq

```python
def f(t, u, ρ=28, σ=10, β=8./3):
    return torch.tensor([σ*(u[1]-u[0]), u[0]*(ρ-u[2])-u[1], u[0]*u[1]-β*u[2]]) 
```

```python
u0 = torch.tensor([0., 1., 1.05])
time_interval = [0. , 2.]
nt = 200
t = torch.linspace(time_interval[0], time_interval[1], nt)
yt = odeint(f, u0, t, rtol=1.e-8, atol=1.e-8, method="dopri5")
```

```python
plt.scatter(yt[:,0], yt[:,1], yt[:,2], c=yt[:,2], cmap='Greens')
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
yt = yt.numpy()
ax.plot3D(yt[:,0], yt[:,1], yt[:,2])
```
